# 부트스트랩이란(Bootstrapping)

## 통계학에서의 부트스트랩(Bootstrap in Statistics)

가설검증(Test)을 하거나 매트릭(metric)을 계산하기 전에 random sampling을 적용하는 방법을 일컫는다.

예를 들어 어떤 집단에서 값을 측정했을때, 그 중에서 임의로 100개를 뽑아서 평균(sample mean)을 구하는 것이다.

부트스르랩은 수집한 데이터의 정확한 확률변수의 분포를 모르고 측정된 통계치의 신뢰도를 가늠할 수 없을 때 사용한다.

측정된 n개의 데이터 중에서 중복을 허용하여 m개를 뽑고, 그들의 평균을 구하기를 여러 번 반복한다. 그럼 평균의 분포를 구할 수 있게 되고, 이로부터 95% 확률로 sample mean이 (a, b) 사이의 구간에 위치한다든가 하는 것을 언급할 수 있다.

이와 같이 분포를 모르는 경우 이외에도 측정된 샘플이 부족한 경우에도 사용할 수 있다.

## ML에서의 부트스트랩(Bootstrap in Machine Learning)

Bootstrapping은 ML에서도 사용되며, 이는 랜덤 샐픔링을 통해 training data를 늘리는 방법이다.

Bootstrapping은 Overfitting을 줄이는데에 도움이된다. 이를 Bagging이라고 한다.

Over-fitting을 줄이기 위해서는 데이터가 많은게 좋다. 하지만 현실적으로 불가능할 때가 많다.

그렇다고 model의 complexity를 줄이는 건 그건 역으로 정확도에 악영향을 줄 수 있다.

만약 계산 시간이 크게 문제가 되지 않는 경우라면, 여러 개의 모델을 써서 model ensemble(여러 모델을 통해 결과를 뽑고, 그 값들을 평균하거나 voting하여 최종 값을 산출하는 방법)을 통해 모델의 안정성을 크게 높일 수 있다. 

각각의 모델이 over-fitting 되어있더라도, 그들을 평균내면 서로서로 상쇄되어 더 general한 모델이 되기 때문이다.

# Bagging 과 Boosting

### Bagging

Bagging은 샘플을 여러 번 뽑아 각 모델을 학습시켜 결과를 집계하는 방법이다.

먼저 대상 데이터로부터 비복원 랜덤 샘플링을 한다. 이렇게 추출한 데이터가 일종의 표본 집단이 된다.

이제 여기에 동일한 모델을 학습시킨다. 그리고 학습된 모델의 예측변수를 집계하여 그 결과로 모델을 생성한다.

대부분의 학습에서는 높은 Bias에 의한 Underfitting, 높은 Variance에 의한 Overfitting이 발생한다.

Bagging과 같은 앙상블 기법은 이러한 오류를 최소화하는데 도움을 준다.

일반적으로 Categorical Data인 경우, 투표 방식 (Voting)으로 집계하며 Continuous Data인 경우, 평균 (Average)으로 집계한다.

대표적인 Bagging 알고리즘으로 RandomForest 모델이 있다. 원래 단일 DecisionTree 모델은 boundary가 discrete 한 모양일 수 밖에 없지만, RandomForest는 여러 트리 모델을 결합하여 이를 넘어수 있다.

### Boosting

Bagging이 일반적인 모델을 만드는데 집중되어 있다면 Boosting은 맞추기 어려운 문제를 맞추는데 초점이 맞춰져 있다.

Boosting도 Bagging과 동일하게 비복원 램덤 샘플링을 하지만, 가중치를 부여한다는 차이점이 있다.

Bagging이 병렬로 학습하는 반면, Boosting은 순차적으로 학습시킨다. 

학습이 끝나면 나온 결과에 따라 가중치가 재분배 된다.

오답에 대해 높은 가중치를 부여하고, 정답에 대해 낮은 가중치를 부여하기 때문에 오답에 더욱 집중할 수 있게 된다.

Boosting기법은 정확도가 높게 나타난다. 하지만, 그만큼 Outlier에 취약하다.

Boosting 알고리즘으로는 AdaBoostm XGBoost, GradientBoost 등이 있다.
